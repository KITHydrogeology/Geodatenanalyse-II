## Frequenzanalyse von Zeitreihen: Motivation und Grundlage

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Übung 1

In einem Beobachtungsbrunnen mit der Bezeichnung BLM-1 im kalifornischen Death Valley wurde eine Zeitreihe des Atmosphäredrucks und des Grundwasserspiegels gemessen. Die Daten befinden sich in der Datei *data/BLM-1.csv*.

**Aufgabe 1**: Lese die Daten ein und bilde beide Zeitreihen ab. Enthalten beide Datensätze sichtbare, zyklische Komponenten?

**Aufgabe 2**: Berechne das Amplitudenspektrum für beide Messreihen. Welche Frequenzen sind dort zu finden und woher stammen sie? *Hinweis*: Das Spektrum muss auf einen Frequenzbereich begrenzt werden.

> Die Daten stammen aus der Publikation von [Rau et al. (2020)](https://hess.copernicus.org/articles/24/6033/2020/)

```python
data = pd.read_csv('data/BLM-1.csv')
data['Datetime[UTC]'] = pd.to_datetime(data['Datetime[UTC]'], dayfirst=True)
data.set_index('Datetime[UTC]', inplace=True)
data
```

```python
fig, axs = plt.subplots(2, figsize=(16, 6))

ax = axs[0]
ax.plot(data.index, data['Baro[m]'])
ax.set_xlabel('Datetime')
ax.set_ylabel('Atmospheric pressure [m]')

ax = axs[1]
ax.plot(data.index, data['BLM-1[m]'])
ax.set_xlabel('Datetime')
ax.set_ylabel('Atmospheric pressure [m]')

plt.show()
```

```python
# Aufgabe 2

fig, axs = plt.subplots(2, 1, figsize=(16,7))

N = len(data)
T = 96
freq = np.fft.fftfreq(N, d=1/T)[0:int(N/2)]

# FFT atmospheric pressure
fft_baro = np.fft.fft(data['Baro[m]'].values)[0:int(N/2)]
amp_baro = (2/N)*np.abs(fft_baro)

idx = (freq > 0.5) & (freq < 2.5)

ax = axs[0]
ax.plot(freq[idx], amp_baro[idx])
ax.set_xlabel('Frequenz (cpd)')
ax.set_ylabel('Ampl. atm. pressure [m]')

# FFT groundwater
fft_gw = np.fft.fft(data['BLM-1[m]'].values)[0:int(N/2)]
amp_gw = (2/N)*np.abs(fft_gw)

ax = axs[1]
ax.plot(freq[idx], amp_gw[idx])
ax.set_xlabel('Frequenz (cpd)')
ax.set_ylabel('Ampl. groundwater [m]')

plt.show()
```
